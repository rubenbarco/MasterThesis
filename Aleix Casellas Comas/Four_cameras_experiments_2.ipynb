```python
def error (reproj, orig):
    diff1 = abs(orig[0][0][0]-reproj[0][0][0])
    diff2 = abs(orig[0][0][1]-reproj[0][0][1])
    return (diff1, diff2)

def relative_error (reproj, orig):
    diff1 = abs(orig[0][0][0]-reproj[0][0][0])
    diff2 = abs(orig[0][0][1]-reproj[0][0][1])
    return (diff1*100/1280, diff2*100/720)

def extrinsic_inverse(M):
    N = np.identity(4)
    N[:3,:4] = M
    return np.linalg.inv(N)[:3,:4]

def extrinsic_mult(A,B):
    M = np.identity(4)
    N = np.identity(4)
    M[:3,:4] = A
    N[:3,:4] = B
    return np.dot(M,N)[:3,:4]
```

```python
def reconstruction_3D(p1, p2, I1, I2, E1, E2, k1, k2):
    projPoints1 = cv2.undistortPoints(p1, I1, k1)
    projPoints2 = cv2.undistortPoints(p2, I2, k2)

    world_coord = cv2.triangulatePoints(E1, E2, projPoints1, projPoints2) 
    world_coord_norm = world_coord/world_coord[3]

    new_world_coord = (world_coord[:-1]/world_coord[3]).reshape(1,1,3)
    return new_world_coord[0]
```

```python
def reprojection_2D(ground_truth1, ground_truth2, ground_truth3, ground_truth4, total_points_3D, I1, I2, I3, I4, E1, E2, E3, E4, k1, k2, k3, k4, length):
    
    error1 = []
    error2 = []
    error3 = []
    error4 = []
    
    for i in range(0,length,3):

        R1 = E1[:3,:3]
        T1 = E1[:,3]
        R2 = E2[:3,:3]
        T2 = E2[:,3]
        R3 = E3[:3,:3]
        T3 = E3[:,3]
        R4 = E4[:3,:3]
        T4 = E4[:,3]

        gt1 = np.array([[[ground_truth1[i], ground_truth1[i+1]]]])
        gt2 = np.array([[[ground_truth2[i], ground_truth2[i+1]]]])
        gt3 = np.array([[[ground_truth3[i], ground_truth3[i+1]]]])
        gt4 = np.array([[[ground_truth4[i], ground_truth4[i+1]]]])
        
        # the length of the total_points_3D is the total length divided by three
        world_coord = total_points_3D[int(i/3)].reshape(1,1,3)
        
        if world_coord.any() != 0: # we just want to reproject if the point 3D is different from (0,0,0). That means that at 
                                   # least two different 2D points are available.
            if gt1.any() != 0:
                reproj_2d_1 = cv2.projectPoints(world_coord, R1, T1, I1, k1)[0]
                error1.append(error(reproj_2d_1, gt1))
            else:
                error1.append([0,0])

            if gt2.any() != 0:
                reproj_2d_2 = cv2.projectPoints(world_coord, R2, T2, I2, k2)[0]
                error2.append(error(reproj_2d_2, gt2))
            else:
                error2.append([0,0])

            if gt3.any() != 0:
                reproj_2d_3 = cv2.projectPoints(world_coord, R3, T3, I3, k3)[0]
                error3.append(error(reproj_2d_3, gt3))
            else:
                error3.append([0,0])

            if gt4.any() != 0:
                reproj_2d_4 = cv2.projectPoints(world_coord, R4, T4, I4, k4)[0]
                error4.append(error(reproj_2d_4, gt4))
            else:
                error4.append([0,0])
        
        else:
            error1.append([0,0])
            error2.append([0,0])
            error3.append([0,0])
            error4.append([0,0])

#     return (np.mean(error1), np.mean(error2), np.mean(error3), np.mean(error4)), number_points

    return error1, error2, error3, error4
```

```python
def sum_error(x,y,c):
    res = np.zeros((number_keypoints,2))
    for i in range (len(x)):
        if y[i] != [0,0]:
            c[i] += 1
        for j in range (len(x[i])):
            res[i][j] = x[i][j]+y[i][j]
    return res, c

def dividing_error(x,c):
    for i in range (len(x)):
        if c[i] != 0:
            x[i] = x[i]/c[i]
    return x

def error_printing(x,y,z,t):
    mean = []
    for i in range (len(x)):
        print("Point {}: [{:.2f},{:.2f}] ; [{:.2f},{:.2f}] ; [{:.2f},{:.2f}] ; [{:.2f},{:.2f}] ---- Mean: {:.2f}".format(i, x[i][0], x[i][1], y[i][0], y[i][1], z[i][0], z[i][1], t[i][0], t[i][1], mean_errors(x[i], y[i], z[i], t[i])))
#         print("{} & [{:.2f},{:.2f}] & [{:.2f},{:.2f}] & [{:.2f},{:.2f}] & [{:.2f},{:.2f}] & {:.2f} \\\\".format(i, x[i][0], x[i][1], y[i][0], y[i][1], z[i][0], z[i][1], t[i][0], t[i][1], mean_errors(x[i], y[i], z[i], t[i])))
        mean.append((x[i], y[i], z[i], t[i], mean_errors(x[i], y[i], z[i], t[i])))
    return mean
        
def mean_errors(a,b,c,d):
    count = 0
    if a.any() != 0:
        count += 1
    if b.any() != 0:
        count += 1
    if c.any() != 0:
        count += 1
    if d.any() != 0:
        count += 1
    if count != 0:
        return (np.mean(a)+np.mean(b)+np.mean(c)+np.mean(d))/count
    else: 
        return 0
```

### Removing highest distances and take mean of the other three


```python
def distance_four(p1, p2, p3, p4, mean):        
    d1 = abs(np.mean(abs(p1) - abs(mean)))
    d2 = abs(np.mean(abs(p2) - abs(mean)))
    d3 = abs(np.mean(abs(p3) - abs(mean)))
    d4 = abs(np.mean(abs(p4) - abs(mean)))
    sum_value = np.array([0,0,0])
    maximum = max(d1,d2,d3,d4)
    if maximum == d1:
        sum_value = sum_value + p2 + p3 + p4
    elif maximum == d2:
        sum_value = sum_value + p1 + p3 + p4
    elif maximum == d3:
        sum_value = sum_value + p1 + p2 + p4
    elif maximum == d4:
        sum_value = sum_value + p1 + p2 + p3
    return sum_value


def distance_five(p1, p2, p3, p4, p5, mean):
    d1 = abs(np.mean(abs(p1) - abs(mean)))
    d2 = abs(np.mean(abs(p2) - abs(mean)))
    d3 = abs(np.mean(abs(p3) - abs(mean)))
    d4 = abs(np.mean(abs(p4) - abs(mean)))
    d5 = abs(np.mean(abs(p5) - abs(mean)))
    sum_value = np.array([0,0,0])
    aux = np.array([d1, d2, d3, d4, d5])
    max1, max2 = np.partition(aux,1)[-2:] # we take the two maximum distances
    if set([d1, d2]) == set([max1,max2]):
        sum_value = sum_value + p3 + p4 + p5
    elif set([d1, d3]) == set([max1,max2]):
        sum_value = sum_value + p2 + p4 + p5
    elif set([d1, d4]) == set([max1,max2]):
        sum_value = sum_value + p2 + p3 + p5
    elif set([d1, d5]) == set([max1,max2]):
        sum_value = sum_value + p2 + p3 + p4
    elif set([d2, d3]) == set([max1,max2]):
        sum_value = sum_value + p1 + p4 + p5
    elif set([d2, d4]) == set([max1,max2]):
        sum_value = sum_value + p1 + p3 + p5
    elif set([d2, d5]) == set([max1,max2]):
        sum_value = sum_value + p1 + p3 + p4
    elif set([d3, d4]) == set([max1,max2]):
        sum_value = sum_value + p1 + p2 + p5
    elif set([d3, d5]) == set([max1,max2]):
        sum_value = sum_value + p1 + p2 + p4
    elif set([d4, d5]) == set([max1,max2]):
        sum_value = sum_value + p1 + p2 + p3
    return sum_value


def distance_six(p1, p2, p3, p4, p5, p6, mean):
    d1 = abs(np.mean(abs(p1) - abs(mean)))
    d2 = abs(np.mean(abs(p2) - abs(mean)))
    d3 = abs(np.mean(abs(p3) - abs(mean)))
    d4 = abs(np.mean(abs(p4) - abs(mean)))
    d5 = abs(np.mean(abs(p5) - abs(mean)))
    d6 = abs(np.mean(abs(p5) - abs(mean)))
    sum_value = np.array([0,0,0])
    aux = np.array([d1, d2, d3, d4, d5, d6])
    max1, max2, max3 = np.partition(aux,1)[-3:] # we take the three maximum distances
    if set([d1, d2, d3]) == set([max1,max2,max3]):
        sum_value = sum_value + p4 + p5 + p6
    elif set([d1, d2, d4]) == set([max1,max2,max3]):
        sum_value = sum_value + p3 + p5 + p6
    elif set([d1, d2, d5]) == set([max1,max2,max3]):
        sum_value = sum_value + p3 + p4 + p6
    elif set([d1, d2, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p3 + p4 + p5
    elif set([d1, d3, d4]) == set([max1,max2,max3]):
        sum_value = sum_value + p2 + p5 + p6
    elif set([d1, d3, d5]) == set([max1,max2,max3]):
        sum_value = sum_value + p2 + p4 + p6
    elif set([d1, d3, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p2 + p4 + p5
    elif set([d1, d4, d5]) == set([max1,max2,max3]):
        sum_value = sum_value + p2 + p3 + p6
    elif set([d1, d4, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p2 + p3 + p5
    elif set([d1, d5, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p2 + p3 + p4
    elif set([d2, d3, d4]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p5 + p6
    elif set([d2, d3, d5]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p4 + p6
    elif set([d2, d3, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p4 + p5
    elif set([d2, d4, d5]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p3 + p6
    elif set([d2, d4, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p3 + p5
    elif set([d2, d5, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p3 + p4
    elif set([d3, d4, d5]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p2 + p6
    elif set([d3, d4, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p2 + p5
    elif set([d3, d5, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p2 + p4
    elif set([d4, d5, d6]) == set([max1,max2,max3]):
        sum_value = sum_value + p1 + p2 + p3
    return sum_value
```

```python
def removing_distance_reconstruction_3D(point1, point2, point3, point4, I1, I2, I3, I4, E1, E2, E3, E4, k1, k2, k3, k4, length):
    mean_points_3D = []
    
    for i in range(0,length,3):
        
        sum_value = np.array([[0,0,0]]) # to sum the values of the views for making the mean
        # I declare the points here to know if I enter to the ifs or not.
        point_3D_12 = np.array([[0,0,0]])
        point_3D_13 = np.array([[0,0,0]])
        point_3D_23 = np.array([[0,0,0]])
        point_3D_14 = np.array([[0,0,0]])
        point_3D_24 = np.array([[0,0,0]])
        point_3D_34 = np.array([[0,0,0]])
        c = 0


        if point1[i] != 0 and point2[i] != 0:
            p1 = np.array([[[point1[i], point1[i+1]]]])
            p2 = np.array([[[point2[i], point2[i+1]]]])
            point_3D_12 = reconstruction_3D(p1, p2, I1, I2, E1, E2, k1, k2)
#             print(point_3D_12, confidence_12)
            c += 1
            sum_value = sum_value + point_3D_12

        if point1[i] != 0 and point3[i] != 0:
            p1 = np.array([[[point1[i], point1[i+1]]]])
            p3 = np.array([[[point3[i], point3[i+1]]]])
            point_3D_13 = reconstruction_3D(p1, p3, I1, I3, E1, E3, k1, k3)
#             print(point_3D_13, confidence_13)
            c += 1
            sum_value = sum_value + point_3D_13
            
        if point2[i] != 0 and point3[i] != 0:
            p2 = np.array([[[point2[i], point2[i+1]]]])
            p3 = np.array([[[point3[i], point3[i+1]]]])
            point_3D_23 = reconstruction_3D(p2, p3, I2, I3, E2, E3, k2, k3)
#             print(point_3D_23, confidence_23)
            c += 1
            sum_value = sum_value + point_3D_23
        
        if point1[i] != 0 and point4[i] != 0:
            p1 = np.array([[[point1[i], point1[i+1]]]])
            p4 = np.array([[[point4[i], point4[i+1]]]])
            point_3D_14 = reconstruction_3D(p1, p4, I1, I4, E1, E4, k1, k4)
#             print(point_3D_14, confidence_14)
            c += 1
            sum_value = sum_value + point_3D_14
        
        if point2[i] != 0 and point4[i] != 0:
            p2 = np.array([[[point2[i], point2[i+1]]]])
            p4 = np.array([[[point4[i], point4[i+1]]]])
            point_3D_24 = reconstruction_3D(p2, p4, I2, I4, E2, E4, k2, k4)
#             print(point_3D_24, confidence_24)
            c += 1
            sum_value = sum_value + point_3D_24
        
        if point3[i] != 0 and point4[i] != 0:
            p3 = np.array([[[point3[i], point3[i+1]]]])
            p4 = np.array([[[point4[i], point4[i+1]]]])
            point_3D_34 = reconstruction_3D(p3, p4, I3, I4, E3, E4, k3, k4)
#             print(point_3D_34, confidence_34)
            c += 1
            sum_value = sum_value + point_3D_34
        
        need_to_operate = 0 # This variable is just to know if we have to operate with confidences or not
        if c != 0:
            if c == 1 or c == 2 or c == 3: # If we just have one, two or three viws, we don't need to take out the one with lowest confidence
                final_mean = sum_value/c
            elif c == 4:
                mean = sum_value/c
                need_to_operate = 4
            elif c == 5:
                mean = sum_value/c
                need_to_operate = 5
            elif c == 6:
                mean = sum_value/c
                need_to_operate = 6
        else:
            final_mean = sum_value

        sum_value = np.array([[0,0,0]])
        if need_to_operate == 4:
            if point_3D_12.any() == 0 and point_3D_13 == 0:
                sum_value = distance_four(point_3D_23, point_3D_14, point_3D_24, point_3D_34, mean)
            elif point_3D_12.any() == 0 and point_3D_23 == 0:
                sum_value = distance_four(point_3D_13, point_3D_14, point_3D_24, point_3D_34, mean)
            elif point_3D_12.any() == 0 and point_3D_14 == 0:
                sum_value = distance_four(point_3D_13, point_3D_23, point_3D_24, point_3D_34, mean)
            elif point_3D_12.any() == 0 and point_3D_24 == 0:
                sum_value = distance_four(point_3D_13, point_3D_23, point_3D_14, point_3D_34, mean)
            elif point_3D_12.any() == 0 and point_3D_34 == 0:
                sum_value = distance_four(point_3D_13, point_3D_23, point_3D_14, point_3D_24, mean)
            elif point_3D_13.any() == 0 and point_3D_23 == 0:
                sum_value = distance_four(point_3D_12, point_3D_14, point_3D_24, point_3D_34, mean)
            elif point_3D_13.any() == 0 and point_3D_14 == 0:
                sum_value = distance_four(point_3D_12, point_3D_23, point_3D_24, point_3D_34, mean)
            elif point_3D_13.any() == 0 and point_3D_24 == 0:
                sum_value = distance_four(point_3D_12, point_3D_23, point_3D_14, point_3D_34, mean)
            elif point_3D_13.any() == 0 and point_3D_34 == 0:
                sum_value = distance_four(point_3D_12, point_3D_23, point_3D_14, point_3D_24, mean)
            elif point_3D_23.any() == 0 and point_3D_14 == 0:
                sum_value = distance_four(point_3D_12, point_3D_13, point_3D_24, point_3D_34, mean)
            elif point_3D_23.any() == 0 and point_3D_24 == 0:
                sum_value = distance_four(point_3D_12, point_3D_13, point_3D_14, point_3D_34, mean)
            elif point_3D_23.any() == 0 and point_3D_34 == 0:
                sum_value = distance_four(point_3D_12, point_3D_13, point_3D_14, point_3D_24, mean)
            elif point_3D_14.any() == 0 and point_3D_24 == 0:
                sum_value = distance_four(point_3D_12, point_3D_13, point_3D_23, point_3D_34, mean)
            elif point_3D_14.any() == 0 and point_3D_34 == 0:
                sum_value = distance_four(point_3D_12, point_3D_13, point_3D_23, point_3D_24, mean)
            elif point_3D_24.any() == 0 and point_3D_34 == 0:
                sum_value = distance_four(point_3D_12, point_3D_13, point_3D_23, point_3D_14, mean)
            final_mean = sum_value/3
        elif need_to_operate == 5:
            if point_3D_12.any() == 0:
                sum_value = distance_five(point_3D_13, point_3D_23, point_3D_14, point_3D_24, point_3D_34, mean)
            elif point_3D_13.any() == 0:
                sum_value = distance_five(point_3D_12, point_3D_23, point_3D_14, point_3D_24, point_3D_34, mean)
            elif point_3D_23.any() == 0:
                sum_value = distance_five(point_3D_12, point_3D_13, point_3D_14, point_3D_24, point_3D_34, mean)
            elif point_3D_14.any() == 0:
                sum_value = distance_five(point_3D_12, point_3D_13, point_3D_23, point_3D_24, point_3D_34, mean)
            elif point_3D_24.any() == 0:
                sum_value = distance_five(point_3D_12, point_3D_13, point_3D_23, point_3D_14, point_3D_34, mean)
            elif point_3D_34.any() == 0:
                sum_value = distance_five(point_3D_12, point_3D_13, point_3D_23, point_3D_14, point_3D_24, mean)
            final_mean = sum_value/3
        elif need_to_operate == 6:
            sum_value = distance_six(point_3D_12, point_3D_13, point_3D_23, point_3D_14, point_3D_24, point_3D_34, mean)
            final_mean = sum_value/3
                    
                    
#         print(final_mean, '\n')
           
        # We always add the points, even if we don't go inside the if. Then we are just adding three 0's.
        mean_points_3D.extend(final_mean)
            
    return mean_points_3D
```
