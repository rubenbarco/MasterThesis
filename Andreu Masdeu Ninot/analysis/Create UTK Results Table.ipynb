```python
import pandas as pd
import pickle
```

```python
task2metric = {'age': 'val_age_loss', 'gender': 'val_gender_acc','ethnicity': 'val_ethnicity_acc' }

# baseline results
tasks_bas = ['age', 'gender', 'ethnicity']
filename = 'history{}agegenderethnicityBaseline.pkl'
direct = 'FirstTake/'
loss_age = []
acc_gen = []
acc_eth = []

final_df = []
for i in range(3):
    results = pickle.load(open(direct+filename.format(i), 'rb'))
    loss_age.append(min(results[task2metric['age']]))
    acc_gen.append(max(results[task2metric['gender']]))
    acc_eth.append(max(results[task2metric['ethnicity']]))

final_df.append(['Baseline Model', sum(loss_age)/3, sum(acc_gen)/3, sum(acc_eth)/3])

# hardparam normal 2t
task2id = {'age': 1, 'gender': 2,'ethnicity': 3 }

for tasks in [['age', 'gender'], ['age', 'ethnicity'], ['gender', 'ethnicity']]:
    
    for shared_blocks in range(1,6):
        
        filename = 'history_hardparam'+tasks[0]+tasks[1]+str(shared_blocks)+'{}.pkl'
        task1 = []
        task2 = []
        line = ['Hard Param Sharing {} shared layers {} {}'.format(shared_blocks, tasks[0], tasks[1]), '-', '-', '-']
        for i in range(3):
            
            results = pickle.load(open(direct+filename.format(i), 'rb'))
            
            
            if tasks[0] == 'age':
                task1.append(min(results[task2metric[tasks[0]]]))
            else:
                task1.append(max(results[task2metric[tasks[0]]]))
                
            task2.append(max(results[task2metric[tasks[1]]]))
        
        line[task2id[tasks[0]]] = sum(task1)/3
        line[task2id[tasks[1]]] = sum(task2)/3
        
        final_df.append(line)
            
#hard param 3t


for shared_blocks in range(1,6):
    filename = 'history_hardparam'+'agegenderethnicity'+str(shared_blocks)+'{}.pkl'
    loss_age = []
    acc_gen = []
    acc_eth = []
    
    for i in range(3):
        results = pickle.load(open(direct+filename.format(i), 'rb'))
        loss_age.append(min(results[task2metric['age']]))
        acc_gen.append(max(results[task2metric['gender']]))
        acc_eth.append(max(results[task2metric['ethnicity']]))

    final_df.append(['Hard Param Sharing {} shared layers 3 tasks'.format(shared_blocks), 
                     sum(loss_age)/3, sum(acc_gen)/3, sum(acc_eth)/3])

#hard param back
for tasks in [['age', 'gender'], ['age', 'ethnicity'], ['gender', 'ethnicity']]:
    
    for shared_blocks in range(1,6):
        
        filename = 'history_hardparam'+tasks[0]+tasks[1]+str(shared_blocks)+'{}back.pkl'
        task1 = []
        task2 = []
        line = ['Hard Param Sharing Backwards {} shared layer {} {}'.format(shared_blocks, tasks[0], tasks[1]), '-', '-', '-']
        for i in range(3):
            
            results = pickle.load(open(direct+filename.format(i), 'rb'))
            
            
            if tasks[0] == 'age':
                task1.append(min(results[task2metric[tasks[0]]]))
            else:
                task1.append(max(results[task2metric[tasks[0]]]))
                
            task2.append(max(results[task2metric[tasks[1]]]))
        
        line[task2id[tasks[0]]] = sum(task1)/3
        line[task2id[tasks[1]]] = sum(task2)/3
    
        final_df.append(line)                   
            
#hard param 3t backwards

loss_age = []
acc_gen = []
acc_eth = []

for shared_blocks in range(1,6):
    filename = 'history_hardparam'+'agegenderethnicity'+str(shared_blocks)+'{}back.pkl'
    loss_age = []
    acc_gen = []
    acc_eth = []
    for i in range(3):
        results = pickle.load(open(direct+filename.format(i), 'rb'))
        loss_age.append(min(results[task2metric['age']]))
        acc_gen.append(max(results[task2metric['gender']]))
        acc_eth.append(max(results[task2metric['ethnicity']]))

    final_df.append(['Hard Param Sharing Backwards {} shared layers 3 tasks'.format(shared_blocks), 
                     sum(loss_age)/3, sum(acc_gen)/3, sum(acc_eth)/3])

    
    
# prop add normal 2t
task2id = {'age': 1, 'gender': 2,'ethnicity': 3 }

for tasks in [['age', 'gender'], ['age', 'ethnicity'], ['gender', 'ethnicity']]:
    
    for shared_blocks in range(1,6):
        
        filename = 'history{}'+tasks[0]+tasks[1]+str(shared_blocks)+'PropAdd.pkl'
        task1 = []
        task2 = []
        line = ['Cross-stitch {} cross-stitch units {} {}'.format(shared_blocks, tasks[0], tasks[1]), '-', '-', '-']
        for i in range(3):
            
            results = pickle.load(open(direct+filename.format(i), 'rb'))
            
            
            if tasks[0] == 'age':
                task1.append(min(results[task2metric[tasks[0]]]))
            else:
                task1.append(max(results[task2metric[tasks[0]]]))
                
            task2.append(max(results[task2metric[tasks[1]]]))
        
        line[task2id[tasks[0]]] = sum(task1)/3
        line[task2id[tasks[1]]] = sum(task2)/3
        
        final_df.append(line)
            
#prop add 3t



for shared_blocks in range(1,6):
    filename = 'history{}'+'agegenderethnicity'+str(shared_blocks)+'PropAdd.pkl'
    loss_age = []
    acc_gen = []
    acc_eth = []
    for i in range(3):
        results = pickle.load(open(direct+filename.format(i), 'rb'))
        loss_age.append(min(results[task2metric['age']]))
        acc_gen.append(max(results[task2metric['gender']]))
        acc_eth.append(max(results[task2metric['ethnicity']]))

    final_df.append(['Cross-stitch {} cross-stitch units 3 tasks'.format(shared_blocks), 
                     sum(loss_age)/3, sum(acc_gen)/3, sum(acc_eth)/3])

#prop add param back
for tasks in [['age', 'gender'], ['age', 'ethnicity'], ['gender', 'ethnicity']]:
    
    for shared_blocks in range(1,6):
        
        filename = 'history{}'+tasks[0]+tasks[1]+str(shared_blocks)+'PropAddb.pkl'
        task1 = []
        task2 = []
        line = ['Cross-stitch Backwards {} cross-stitch units {} {}'.format(shared_blocks, tasks[0], tasks[1]), '-', '-', '-']
        for i in range(3):
            
            results = pickle.load(open(direct+filename.format(i), 'rb'))
            
            
            if tasks[0] == 'age':
                task1.append(min(results[task2metric[tasks[0]]]))
            else:
                task1.append(max(results[task2metric[tasks[0]]]))
                
            task2.append(max(results[task2metric[tasks[1]]]))
        
        line[task2id[tasks[0]]] = sum(task1)/3
        line[task2id[tasks[1]]] = sum(task2)/3
    
        final_df.append(line)                  
            
#prop param 3t backwards



for shared_blocks in range(1,6):
    filename = 'history{}'+'agegenderethnicity'+str(shared_blocks)+'PropAddback.pkl'
    loss_age = []
    acc_gen = []
    acc_eth = []
    for i in range(3):
        results = pickle.load(open(direct+filename.format(i), 'rb'))
        loss_age.append(min(results[task2metric['age']]))
        acc_gen.append(max(results[task2metric['gender']]))
        acc_eth.append(max(results[task2metric['ethnicity']]))

    final_df.append(['Cross-stitch Backwards {} cross-stitch units 3 tasks'.format(shared_blocks), 
                     sum(loss_age)/3, sum(acc_gen)/3, sum(acc_eth)/3])
```

```python
df = pd.DataFrame(final_df, columns=['Model', 'Age', 'Gender', 'Ethnicity'])
```
